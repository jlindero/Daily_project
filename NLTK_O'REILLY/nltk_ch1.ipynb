```
def lexical_diversity(text):
  return len(text) / len(set(text)) 
def percentage(count, total):
  return 100 * count / total
```

```
print("text used about ",lexical_diversity(text3), 'times in text 3')
print("text",lexical_diversity(text5), 'times in text 5')
print("percent",percentage(4, 5))
print("percent a used in text4", percentage(text4.count('a'), len(text4)))
```
